```python
from bokeh.io import output_notebook, show
from bokeh.models import HoverTool,BoxZoomTool,WheelZoomTool,ResetTool
import pandas as pd
output_notebook()

import math,numpy as np

def voltage2temp( voltage ):
    resistor = (voltage/10)/(3.27 - voltage/1000)
    temp = 1/(np.log(resistor/47)/4050 +1/(25+273.15))-273.15
    #from IPython.core.debugger import Tracer;Tracer()()
    if temp > 0:
        temperature = (temp * 100 + 5)/100
    else:
        temperature = (temp * 100 - 5)/100
    return temperature

def voltage2temp_env( d , rh):#mv
    va = 3.3*d/255
    vcc = 3.3
    b = 4050
    ro = 47
    to = 25 
    #Tc=1/(ln(Va*Rh/(Vcc-Va)/R0)/B+1/(Tc0+273.15))-273.15
    temp = 1/(np.log(va*rh/(vcc-va)/ro)/b+1/(to+273.15))-273.15
    #from IPython.core.debugger import Tracer;Tracer()()
    return temp

def d2voltage(d):
    return 3.3*d/255

src=pd.DataFrame({'voltage':np.arange(0,255,1)})
X = src['voltage'].values
Y1 = [ voltage2temp_env(x,30) for x in X ]
Y2 = [ voltage2temp_env(x,3.3) for x in X ]
Y3 = [ d2voltage(x) for x in X]

def abs_hex(v):
    if v > 0:
       return hex(int(v))
    else:
       return hex(0)

def print_table(values):
    table = [ abs_hex(v) for v in values]
    print table

from bokeh.plotting import figure

hover = HoverTool(
     tooltips=[
            ("index", "$in"),
            ("(x,y)", "($x, $y)"),
            ("desc", "@desc"),
        ]
)

p = figure(height=300,title='(Python)',tools=[hover,BoxZoomTool(),WheelZoomTool(),ResetTool()])
p.title_text_font = 'PilGi'
p.line(x=X,y=Y1,line_width=1,color='red',alpha=.3)
p.line(x=X,y=Y2,line_width=1,color='blue',alpha=.3)
p.line(x=X,y=Y3,line_width=1,color='green',alpha=.3)
show(p)
```
